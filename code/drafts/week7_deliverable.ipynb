```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
 
import warnings
warnings.filterwarnings("ignore")

# show float in two decimal form
plt.style.use('ggplot')
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

```python
VIC = pd.read_csv('../../data/vic.csv')
data = VIC[(VIC['time'] >= "2020-07-17 00:00:00") & (VIC['time'] < "2020-07-18 00:00:00")].reset_index(drop=True)
data.sort_values(['spot_price'])
```

```python
plt.figure(figsize=(35,7))
x = data.period
y = data.spot_prices
plt.plot(x, y)
plt.xticks(data.period, fontsize=20)
k = 20
maxi = data.sort_values(['spot_price']).tail(k).reset_index(drop=True)
for i in range(k):
    plt.annotate(maxi.loc[i].spot_price, xy=(maxi.loc[i].period,maxi.loc[i].spot_price), fontsize=10)
    
k = 20
mini = data.sort_values(['spot_price']).head(k).reset_index(drop=True)
for i in range(k):
    plt.annotate(mini.loc[i].spot_price, xy=(mini.loc[i].period,mini.loc[i].spot_price), c='b', fontsize=10)

plt.show()
```
